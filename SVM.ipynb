```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
```

```python
digit.keys()
```

```python
print(digit["DESCR"])
```

```python
digits=pd.DataFrame(digit["data"],columns=digit["feature_names"])
```

```python
digits
```

```python
digits.describe()
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
digits.columns
```

```python
x=digits[['pixel_0_0', 'pixel_0_1', 'pixel_0_2', 'pixel_0_3', 'pixel_0_4',
       'pixel_0_5', 'pixel_0_6', 'pixel_0_7', 'pixel_1_0', 'pixel_1_1',
       'pixel_1_2', 'pixel_1_3', 'pixel_1_4', 'pixel_1_5', 'pixel_1_6',
       'pixel_1_7', 'pixel_2_0', 'pixel_2_1', 'pixel_2_2', 'pixel_2_3',
       'pixel_2_4', 'pixel_2_5', 'pixel_2_6', 'pixel_2_7', 'pixel_3_0',
       'pixel_3_1', 'pixel_3_2', 'pixel_3_3', 'pixel_3_4', 'pixel_3_5',
       'pixel_3_6', 'pixel_3_7', 'pixel_4_0', 'pixel_4_1', 'pixel_4_2',
       'pixel_4_3', 'pixel_4_4', 'pixel_4_5', 'pixel_4_6', 'pixel_4_7',
       'pixel_5_0', 'pixel_5_1', 'pixel_5_2', 'pixel_5_3', 'pixel_5_4',
       'pixel_5_5', 'pixel_5_6', 'pixel_5_7', 'pixel_6_0', 'pixel_6_1',
       'pixel_6_2', 'pixel_6_3', 'pixel_6_4', 'pixel_6_5', 'pixel_6_6',
       'pixel_6_7', 'pixel_7_0', 'pixel_7_1', 'pixel_7_2', 'pixel_7_3',
       'pixel_7_4', 'pixel_7_5', 'pixel_7_6']]
y=digits['pixel_7_7']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
svm.fit(X_train,y_train)
```

```python
ypred=svm.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
cl=classification_report(y_test,ypred,output_dict=True)
```

```python
cl1=pd.DataFrame(cl).transpose()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(ypred,y_test,color="r")
plt.plot(ypred,y_test,color='b',lw=3)
```

```python
cm=confusion_matrix(ypred,y_test,)
```

```python
cm1=pd.DataFrame(cm)
```

```python
cm_display=ConfusionMatrixDisplay(confusion_matrix=cm)
cm_display.plot()
```

```python
grid_predictions = grid.predict(X_test)
```
